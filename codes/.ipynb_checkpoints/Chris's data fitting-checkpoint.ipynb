```julia
#Spatially explicit metapopulation version of the model

using CSV, DataFrames, Random, LinearAlgebra, Distances, Distributions, SpecialFunctions, Plots


#Calculate the probabilities of getting an offspring with trait k when the gametes with 'trait 
#value' i an j are combined

function qgprob(n::Int64)
    
    #All possible phenotypes
    pheno= collect(1:(2*n+1)) ./ (2*n+1)
    nt=length(pheno)

    G=zeros(Float64,n+1,n+1,n+1)

    for i in 0:n, j in 0:i, k in max(0,(i+j-n)):min(n,(i+j))
                m=collect(0:min(j,k,i+j-k))
                G[1+i,1+j,1+k]=sum(pdf.(Hypergeometric(i,n-i,j),m).*pdf.(Binomial.(i+j .- (2 .* m)),k .- m))
    end

    for k in 0:n
        G[:,:,1+k]=G[:,:,1+k]+transpose(G[:,:,1+k])
        for i1 in 0:n
            G[i1+1,i1+1,k+1] /= 2
        end
    end

    ind_haplR=zeros(Float64,2*n+1, 2*n+1)

    for k in 0:n
        for i in 0:n
             ind_haplR[1+i,1+k] = G[1+i,1,1+k]
            for j in 0:n
                ind_haplR[1+j+n,1+k]=G[1+n,1+j,1+k]
            end
        end
    end

    R=zeros(Float64,nt,nt,nt)

    for i in 0:(2*n), j in 0:(2*n), q in 0:(2*n)
         R[1+i,1+j,1+q]= sum(ind_haplR[1+i,1 .+ (0:q)] .* 
                             ind_haplR[1+j,1+q .- (0:q)])
    end
        
    return R
end
    
    
#Function to simulate population dynamics
    
function multipop(Ns::Vector{Float64},r::Vector{Float64},K::Vector{Float64},R::Array{Float64},
                    mmat::Matrix{Float64},SST::Matrix{Float64})
        
    tsteps=size(SST)[2]
        
    nloci=(size(R)[1]-1)/2
    pheno= collect(1:(2*nloci+1)) ./ (2*nloci+1)
        
    N=rand(Uniform(0,1.0),length(Ns),size(R)[1])
    Ng0= N ./ sum(eachcol(N))
    Np0= Ng0 .* Ns 
    Ngen=deepcopy(Ng0)
    Np=deepcopy(Np0)

    #Set up optimal traits value change with time using SST data
    #At time=0, all population have Topt at 0.5.
        
    topts=0.5 .+ (0.8.* SST)

    #precalculate selection pressure on phenotypes across time steps
    probsurv=zeros(Float64, tsteps, length(Ns), length(pheno))
        
    [probsurv[x,y,:]=exp.(-((abs.(topts[y,x] .- pheno)) .^2) ./ 0.16) for x in 1:tsteps, y in 1:length(Ns) ]
     
    #Store results
    res=zeros(Float64,tsteps+1,length(Ns),length(pheno))

    res[1,:,:]=Np0

    for t in 1:tsteps
            
        if sum(floor.(Np))<1
                break
        
        else

            #Selection event
            
            Np=Np .* (probsurv[t,:,:])

            #Reproduction event
                
                newgen=zeros(Float64,length(Ns),length(pheno))

                for i in 1:size(Ngen)[1]
                    
                    Ngen[i,:]=Np[i,:]/sum(Np[i,:])

                    probs=Ngen[i,:]*(Ngen[i,:]')

                    for j in 1:size(Ngen)[2]

                        newgen[i,j]=sum(probs.*R[:,:,j])
                    end

                    Np[i,:]=sum(Np[i,:]) .* newgen[i,:]

                end
            

            #Settlement

            for i in 1:size(Np)[1]
                    
                #Logistic growth to simulate crowding effect during settlement
                Ngen[i,:]=Np[i,:]/sum(Np[i,:])
                                                        
                Np[i,:]=(sum(Np[i,:])+ (r[i]*sum(Np[i,:])*(1-(sum(Np[i,:])/K[i])))) .* Ngen[i,:]

            end

            #Migration
            #add fraction of populations to each other (uniformly across all phenotypes) according to
            #the migration matrix mmat
            for i in 1:size(Np)[1], j in 1:size(Np)[1]
    
                if i != j
                    Np[i,:] += mmat[i,j] .* Np[j,:]
                end
    
                end
    
                Np[Np .<1 ] .= 0
    
                res[t+1,:,:]=Np

        end
    end
    
    return res
        
end

moving_average(vs,n) = [sum(@view vs[i:min((i+n-1),length(vs))])/n for i in 1:(length(vs)-(n-1))]

```

```julia
#Model parameters/Function inputs
SSTs=DataFrame(rcp2_6=collect(range(start=0,stop=0.775, length=95)),
              rcp4_5=collect(range(start=0,stop=1.6, length=95)),
              rcp6_0=collect(range(start=0,stop=1.925, length=95)),
              rcp8_5=collect(range(start=0,stop=3.0, length=95)))

tsteps=size(SSTs)[1]
       
SSTdat=zeros(Float64,11,tsteps,4)
for i in 1:9
    SSTdat[i,:,:] .= SSTs .- ((i-1)/2)
end
SSTdat[10,:,:] .= SSTs .- 1.0
SSTdat[11,:,:] .= SSTs .- 1.0

SSTdat += rand(Normal(0,0.1), 11,tsteps,4);
#Temperature rise data
#At time=0, 1st population is at the highest temp (O). 8 northern pops. have gradually decreasing temps.
#pops 10 and 11 are both at -1 temp. Plus noise obv.


#Assume 6 loci.
R=qgprob(6)

#11 populations

#(9 connected populations, 2 isolated ones. 1 of 9 is the source pop from south)
Ns=[10000.0,1000,1000,1000,1000,1000,1000,1000,1000]

#Growth rate
r=repeat([1.03],length(Ns))

#Carrying capacity
K=12000 .*ones(Float64,9)

#Migration matrix NpopsXNpops: Indicates the proportion of the population migrating
#from pop i to pop j
#Use pop. geo. coordinates
points=[16.50 16.87 20.14 22.03 22.20 22.26 24.00 26.26 27.36;
    42.10 41.79 40.26 38.45 38.51 39.00 37.90 36.43 35.66]
mmat=pairwise(Euclidean(),points)
mmat=0.5*pdf(Normal(0.0,10),mmat)/pdf(Normal(0.0,10),0)

# Params to vary: rep,RCP scenarios 

reps=collect(1:5)

rcps=collect(1:4)


pars=collect(Iterators.product(reps,rcps))

```

```julia


SST=SSTdat[:,:,1]
res=multipop(Ns,r,K,R,mmat,SSTdat[:,:,1])
popdat=zeros(Float64,size(res)[1],size(res)[2])
    
for i in 1:size(res)[1]
    popdat[i,:]=sum(res[i,:,:],dims=2)
end

p1=plot(plot(moving_average(sum(popdat,dims=2),10)),label="RCP 2.6",ticks=false,xlabel="Time",ylabel="Population size")

labels=["RCP 4.5","RCP 6.0","RCP 8.5"]

for i in 2:4
    
    SST=SSTdat[:,:,i]
    res=multipop(Ns,r,K,R,mmat,SSTdat[:,:,4])
    
    popdat=zeros(Float64,size(res)[1],size(res)[2])
    
    for i in 1:size(res)[1]
        popdat[i,:]=sum(res[i,:,:],dims=2)
    end

    plot!(moving_average(sum(popdat,dims=2),10),label=labels[i-1])
end
    
p1
```
